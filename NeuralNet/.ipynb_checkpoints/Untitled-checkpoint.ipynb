```python
#category non-numerical data
#using mean to represent NA data
trainNum = train.shape[0]
oeDF = train.copy()

for column in train:    
    if train[column].dtype == 'object':
        oeDF[column] = oeDF[column].astype('category').cat.codes
    else:
        temp = train.iloc[:trainNum-1, column].fillna(train[column].mean())
        mean = temp.mean()
        std = temp.std()
        oeDF[column] = (oeDF[column] - mean) /std
indexedTrain = oeDF.head(train.shape[0])
```

```python
trainNum = indexedTrain.shape[0]
for column in train:
    mean = train[column].mean()
    std = train[column].std()
    oeDF[column] = (oeDF[column] - mean) /std
```

```python
ncols = len(indexedTrain.columns)
nrows = len(indexedTrain.index)
```

```python
indexedTrain.X = indexedTrain.iloc[:, 0:(ncols -1)].values.reshape(nrows, ncols-1)
indexedTrain.y = indexedTrain.iloc[:, (ncols-1)].values.reshape(nrows, 1)
```

```python
input_layer_size = len(indexedTrain.X[0])
if not isinstance(indexedTrain.y[0], np.ndarray):
    output_layer_size = 1
else:
    output_layer_size = len(indexedTrain.y[0])
```

```python
h1=4
h2=2
w01 = 2 * np.random.random((input_layer_size, h1)) - 1
X01 = indexedTrain.X
delta01 = np.zeros((input_layer_size, h1))
w12 = 2 * np.random.random((h1, h2)) - 1
X12 = np.zeros((len(indexedTrain.X), h1))
delta12 = np.zeros((h1, h2))
w23 = 2 * np.random.random((h2, output_layer_size)) - 1
X23 = np.zeros((len(indexedTrain.X), h2))
delta23 = np.zeros((h2, output_layer_size))
deltaOut = np.zeros((output_layer_size, 1))
```

```python
def activation(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_derivative(x):
        return x * (1 - x)
```

```python
in1 = np.dot(indexedTrain.X, w01)   
X12 = activation(in1)
in2 = np.dot(X12, w12)
X23 = activation(in2)
in3 = np.dot(X23, w23)
out = activation(in3)
```

```python
def compute_output_delta(out):
    delta_output = (indexedTrain.y - out) * (sigmoid_derivative(out))
    deltaOut = delta_output
def compute_hidden_layer2_delta():
    delta_hidden_layer2 = (deltaOut.dot(w23.T)) * (sigmoid_derivative(X23))
    delta23 = delta_hidden_layer2

def compute_hidden_layer1_delta():
    delta_hidden_layer1 = (delta23.dot(w12.T)) * (sigmoid_derivative(X12))
    delta12 = delta_hidden_layer1

def compute_input_layer_delta():
    delta_input_layer = np.multiply(sigmoid_derivative(X01), delta01.dot(w01.T))
    delta01 = delta_input_layer
```

```python
compute_output_delta(out)
compute_hidden_layer2_delta()
compute_hidden_layer1_delta()
```
